Среднее значение сичтется долше, но ненадолго

```python
%%time
education = list(df.reset_index().groupby(by="education")['writing score'])
for j in range(len(education)):
    min_val = list(education[j][1])[0]
    for i in range(len(list(education[j][1]))):
        if i < len(list(education[j][1])) - 1:
            if min_val > list(education[j][1])[i+1]:
                min_val = list(education[j][1])[i+1]
    print(education[j][0]+": ", min_val)
```

#### Задание 2.15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
def checking(df):
    if df["Math_PassStatus"] == "F" or df["Reading_PassStatus"] == "F" or df["Writing_PassStatus"] == "F":
        return "F"
    else:
        return "P"
df["OverAll_PassStatus"] = df[["Math_PassStatus", "Reading_PassStatus", "Writing_PassStatus"]].apply(lambda x: checking(x), axis=1)
df
```
